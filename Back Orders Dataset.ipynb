# Problem - Is to identify products at risk of backorder before the event occurs so the business has time to react.
## Data
## Data file contains the historical data for the 8 weeks prior to the week we are trying to predict. The data was taken as weekly snapshots at the start of each week. Columns are defined as follows:

* sku - Random ID for the product


* national_inv - Current inventory level for the part


* lead_time - Transit time for product (if available)


* in_transit_qty - Amount of product in transit from source


* forecast_3_month - Forecast sales for the next 3 months


* forecast_6_month - Forecast sales for the next 6 months


* forecast_9_month - Forecast sales for the next 9 months


* sales_1_month - Sales quantity for the prior 1 month time period


* sales_3_month - Sales quantity for the prior 3 month time period


* sales_6_month - Sales quantity for the prior 6 month time period


* sales_9_month - Sales quantity for the prior 9 month time period


* min_bank - Minimum recommend amount to stock


* potential_issue - Source issue for part identified


* pieces_past_due - Parts overdue from source


* perf_6_month_avg - Source performance for prior 6 month period


* perf_12_month_avg - Source performance for prior 12 month period


* local_bo_qty - Amount of stock orders overdue


* deck_risk - Part risk flag


* oe_constraint - Part risk flag


* ppap_risk - Part risk flag


* stop_auto_buy - Part risk flag


* rev_stop - Part risk flag


* went_on_backorder - Product actually went on backorder. This is the target value.


* Yes or 1 : Product backordered


* No or 0  : Product not backordered

```python
back_new = back_orders.dropna()
back_new.head()
```

```python
back_new.plot(kind='box')
plt.xticks(rotation=90)
plt.show()
```

```python
feature = back_new.drop('went_on_backorder', axis = 1)

num = feature.select_dtypes(include = [np.number])

cat = feature.select_dtypes(include = [np.object])

dummy_var = pd.get_dummies(data = cat, drop_first = True)

x = pd.concat([num, dummy_var], axis = 1)
x = x.drop('sku',axis=1)

x.head()
```

```python
target = back_new.went_on_backorder
target = target.map({'Yes':1,'No':0})
target = target.astype('int64')
```

```python
x1 = x
y = target

xtrain , xtest , ytrain , ytest = train_test_split(x1,y,random_state=10,test_size=0.2)

decision_tree = DecisionTreeClassifier(random_state=10,max_depth=5)

decision_tree = decision_tree.fit(xtrain,ytrain)

ypred = decision_tree.predict(xtest)

cr = metrics.classification_report(ytest,ypred)

print('Classification Report of DT Model is :\n',cr)
```

```python
important_features = pd.DataFrame({'Features':xtrain.columns,'Importance':decision_tree.feature_importances_})

important_features.sort_values('Importance',ascending=False)
```

```python
from sklearn import tree
plt.figure(figsize=(100,80))
decision_tree_classification = DecisionTreeClassifier(max_depth=5)

dt_model = decision_tree_classification.fit(xtrain,ytrain)

labels = xtrain.columns

dot_data = tree.plot_tree(dt_model,filled=True,feature_names=labels,class_names=['No','Yes'])
```

```python
knn_classification = KNeighborsClassifier(n_neighbors=3,metric='euclidean')

knn_model = knn_classification.fit(xtrain,ytrain)

ypred = knn_model.predict(xtest)

cr = metrics.classification_report(ytest,ypred)

print('Classification Report of KNN Model is :\n',cr)
```

```python
from sklearn.ensemble import RandomForestClassifier

rf_classification = RandomForestClassifier(n_estimators=500,random_state=10,max_depth=10,max_features=10)

rf_model = rf_classification.fit(xtrain,ytrain)

ypred = rf_model.predict(xtest)

cr = metrics.classification_report(ytest,ypred)

print('Classification Report of RF Model is :\n',cr)
```

```python
from sklearn.naive_bayes import GaussianNB

gnb = GaussianNB()

gnb_model = gnb.fit(xtrain,ytrain)

ypred = gnb_model.predict(xtest)

cr = metrics.classification_report(ytest,ypred)

print('Classification Report of NB Model is :\n',cr)
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

mylist = []

for i in range(x1.shape[1]):
    mylist.append(variance_inflation_factor(x1.values,i))
    
pd.DataFrame({"Features":x1.columns,'VIF':mylist}).sort_values("VIF",ascending=False)
```

```python
x2 = x1.drop(['sales_6_month','sales_9_month','forecast_6_month','forecast_9_month'],axis=1)
mylist = []

for i in range(x2.shape[1]):
    mylist.append(variance_inflation_factor(x2.values,i))
    
pd.DataFrame({"Features":x2.columns,'VIF':mylist}).sort_values("VIF",ascending=False)
```

```python
from sklearn.linear_model import LogisticRegression

y = target

xtrain , xtest , ytrain , ytest = train_test_split(x2,y,random_state=10,test_size=0.2)

logreg = sm.Logit(ytrain,xtrain).fit()

ypred_prob = logreg.predict(xtest)

ypred = [ 0 if x < 0.5 else 1 for x in ypred_prob]

cr = metrics.classification_report(ytest,ypred)

print('Accuracy of Model is :\n',cr)
```

```python
fpr, tpr, thresholds = roc_curve(ytest,ypred_prob)

youdens_table = pd.DataFrame({'TPR':tpr, 'FPR':fpr,'Threshold':thresholds})

youdens_table['Difference'] = youdens_table.TPR - youdens_table.FPR

youdens_table = youdens_table.sort_values('Difference',ascending=False).reset_index(drop=True)

youdens_table.style.highlight_max(color = 'yellow', subset = ['Difference'])
```
